# Action1: 二分查找

```python
# 1.定义值 0.0000000001
epsion = 0.0000000001

#2.编写函数
def sqrt10():
    # 最大值，最小值
    low,high = 3,4 
    
    #中间值
    mid = (low+high)/2 
    # 第二种方法
#     mid = low+(low+high)/2  # 3.162277660161635

    while  high-low > epsion:
        if mid * mid > 10:
            # 在左侧进行二分查找
            high = mid
        else:
            low = mid
        mid = (low+high) / 2
    return mid

# 调用函数输出结果
print("Result:",sqrt10())


import numpy as np
print("np.sqrt(10):",np.sqrt(10))
```
